```python
#найти рестораны, открытые в октябре
df_new_time = df.copy()
df_new_time['dateAdded'] = pd.to_datetime(df_new_time['dateAdded'])
df_new_time['dateUpdated'] = pd.to_datetime(df_new_time['dateUpdated'])

df_new_time[df_new_time['dateAdded'].dt.month == 10][:5].index.tolist()
```

```python
#Сделайте группировку по месяцу открытия ресторана (переменная dateAdded).
#Рассчитайте по количеству уникальных id ресторанов в каждом месяце. 
#Значения dateAdded должны при этом быть индексами
df_new_time.groupby(df_new_time['dateAdded'].dt.month).agg({'id':'nunique'})
```

```python
# В этом задании нужно добавить столбец update_deltaс количеством целых дней,
#полученных от разницы между dateUpdated и dateAdded
df_new_time = df_new_time.assign(update_delta = (df_new_time['dateUpdated'] - df_new_time['dateAdded']).dt.days)
```

```python
#для каждого города найдите (в указанном порядке):
#- среднее по переменной update_delta
#- широту самой северной закусочной (чем севернее расположена закусочная, тем больше будет значение широты)
df_new_time.groupby(df_new_time['city'],as_index=False).agg({'update_delta':'mean', 'latitude': 'max'})
```

```python
# Посчитайте среднее и медиану update_delta = (df_new_time['dateUpdated'] - df_new_time['dateAdded']
#Перечислите значения через пробел (сначала среднее, затем медиану), округлив до 2 знаков
df_new_time['update_delta'].median()
df_new_time['update_delta'].mean()
```

```python
zep_mean = df_new_time.groupby(df_new_time['city'])['update_delta'].mean()['Zephyrhills']
zep_mean
```

```python
#Отобрать записи, у которых более 20 категорий, 
#Cгруппировать по провинциям и подсчитать минимальную longitude в каждой. 
#Округлите значения longitude до 3-х знаков после запятой.
#Cохранить результирующий датафрейм в файл в формате csv с сепаратором ;. 
#Сам датафрейм должен содержать две колонки: province и longitude

#df_filter_cat = df[(df['categories'].str.split(',')).str.len() > 20]
df_filter_cat = df[(df['categories'].str.split(',')).apply(lambda x: len(x)) > 20]
res = df_filter_cat.groupby(df_filter_cat['province'])['longitude'].min().round(3)
res.to_csv('df_filter_cat.csv', sep=';')
```
